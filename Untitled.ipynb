```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
from sklearn.linear_model import LinearRegression
df = pd.read_csv('dengue.csv')
```

```python
df.head
```

```python
for column in df.columns:
    df[column].fillna(df[column].mode()[0], inplace=True)
```

```python
df.isnull()
```

```python
from sklearn import preprocessing
label_encoder = preprocessing.LabelEncoder()
```

```python
#DATASET YANGT ADA ADALAH DATA DENGAN VALUE YES OR NO, UNTUK DAPAT MELAKUKAN PEMROSESAN TERHADAP DATA TERSEBUT
#PERLU DILAKUKAN ENCODING UNTUK MENGHASILKANA DATA NUMERIC DENGAN YES = 1 NO = 0

for column in df.columns:
    df[column] = label_encoder.fit_transform(df[column])
```

```python
#UNTUK MENGETAHUI KORELASI ANTARA DATA YANG ADA DENGAN KEMUNGKINAN MENGIDAP DENGUE
#KORELASI ANTARAB DATA DAPAT DILIHAT PADA KOLOM DIBAWAH, SEMAKIN MENDEKATI ANGKA 1 MAKA SEMAKIN TINGGI KORELASINYA

corr = df.corr()
corr.style.background_gradient(cmap='coolwarm')
corr.style.background_gradient(cmap='coolwarm').set_precision(2)
```

```python
X = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
col_=df.columns.tolist()[2:]
for i in df.columns.tolist()[2:]:
    sns.lmplot(x=i,y='dengue',data=df,markers='.')
```

```python
x=df[col_].drop('dengue',1)
y=df['dengue']
```

```python
sc=StandardScaler()
```

```python
x_sc=sc.fit_transform(x)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_sc,y,test_size = 0.3 , random_state = 50 )
```

```python
lr=LinearRegression()
lr_model=lr.fit(x_train, y_train)
```

```python
y_pre=lr_model.predict(x_test)
```

```python
rmse=np.sqrt(mean_squared_error(y_test, y_pre))
print("error RMSE adalah", rmse)
```
